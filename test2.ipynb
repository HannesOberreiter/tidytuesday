```python
mtcars = sm.datasets.get_rdataset("mtcars", "datasets", cache=True).data
data = pd.DataFrame(mtcars)
data = data.assign(carb_2 = data["carb"])

pd.concat(
    [
        # create df without the carb and carb_2 cols
        data.loc[:, ~data.columns.isin(['carb', 'carb_2'])], 
        # pivot to wide format
        data.pivot_table(values='carb', index=data.index, columns='carb_2')
    ], 
    axis=1
    )
```
